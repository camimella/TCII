La banda de paso se encuentra en las frecuencias bajas, y ya que la función es simétrica, el centro de la misma corresponde a s=0. Observando S11, vemos que en ese punto no se llega a 0 (es un valor cercano), con lo cual no está totalmente adaptado. Por otro lado, el centro de la banda de detenida está en infinito, y S11 en infinito es 1, es decir que es un circuito abierto. En este punto, el circuito se desadaptó completamente.
